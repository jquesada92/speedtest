```python
import pandas as pd 
df = pd.read_parquet('data')
```

```python
df.sort_values('timestamp',ascending= False).head()
```

```python
def mean_rollin_intervals(df,col):
    WINDOW = 4
    _df = df.copy()
   
    _df.set_index('timestamp',inplace=True)
    _df.sort_index(inplace = True )
    series = _df[col]
    mean = series.rolling(WINDOW).mean()
    std = series.rolling(WINDOW).std()

    lower_bound = mean - 1.96*std
    upper_bound = mean + 1.96*std
    return (pd.concat([
                    lower_bound.rename('lower'),
                     mean.rename('avg'),
                     upper_bound.rename('upper'),
                     std.rename('std')],axis = 1))
```

```python
download_df = mean_rollin_intervals(df,"Download_Mbps")
```

```python
import plotly.graph_objects as go
```

```python
download_df.columns
```

```python
def historical_line_chart(df):
    x = df.index


    fig = go.Figure()
    line={
                "color":"rgba(171,157,242,0)"
            }

        
    fig.add_trace(
     go.Scatter(x=x,y=download_df['upper'],mode='lines', fill=None,line=line,name=None)

    )
    fig.add_trace(
     go.Scatter(x=x,y=download_df['avg'],name="Download (Mbps)",line = {
                "color":"rgb(171,157,242)"
            }, fill='tonexty')

    )
    fig.add_trace(
     go.Scatter(x=x,y=download_df['lower'], fill='tonexty',mode='lines',line=line,name=None)

    )
    
    return fig

historical_line_chart(download_df).show()
```
